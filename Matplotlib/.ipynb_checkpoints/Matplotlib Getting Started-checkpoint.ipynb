## Importing matplotlib and other libraries

```python
import numpy as np

import pandas as pd

import matplotlib
```

## Checking Matplotlib version

```python

print(matplotlib.__version__)
```

## Importing matplotlib under plt alias


```python

import matplotlib.pyplot as plt
```

## Drawing line from position (0, 0) to (6, 250
                                     )

```python

x_point = np.array([0, 6])

y_point = np.array([0, 250])

plt.plot(x_point, y_point) # plot the graph

plt.show() #displays the graph
```


-> To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.


```python

plt.plot(x_point, y_point, 'o' )

plt.show()
```

## Plotting multiple points

```python

x_points = np.array([1, 2, 6, 8])

y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points)

plt.show()
```

## Markers


-> marker emphasizes on each point with a specified marker.


```python

plt.plot(x_points, y_points, marker = '*')

plt.show()
```

## Format Strings fmt



-> You can use also use the shortcut string notation parameter to specify the marker.


->This parameter is also called fmt, and is written with this syntax:
    

marker|line|color


```python

plt.plot(x_points, y_points, 'd:b') # diamond with blue dotted lines

plt.show()
```

```python

plt.plot(x_points, y_points, 'v--m') # triangle down with magenta color  lines

plt.show()
```

## Marker Size



You can use the keyword argument markersize or the shorter version, 

ms to set the size of the markers:

```python
y = np.array([1, 5, 2, 4, 3, 7])

plt.plot(y, 's-.g', ms = 20) #the size of marker is magnified.

plt.show()
```

## Marker Color


You can use the keyword argument markeredgecolor or 

the shorter mec to set the color of the edge of the markers:

```python
plt.plot(y, 's-c', ms = 5, mec = 'b') # marker with blue color

plt.show()
```

## Matplotlib Line

### Linestyle

-> The keyword linestyle, or shorter ls, can be used to change the style of the plotted line.

```python

plt.plot(y, linestyle = 'dotted')


plt.show()
```

-> Other linestyles include:
    
    1. dashed or '-'
    
    2. dotted or ':'
    
    3.solid or '--'
    
    4.dashdot or '-.'
    
    5. None or "or"

### Line color

-> The keyword argument color or c, is used to set the color of the line

```python
y2 = np.array([2.1, 4, 3.5, 0.9, 4.1])

plt.plot(y, linestyle = 'dotted', color = 'hotpink')

plt.plot(y2, linestyle = 'dashed', color = 'k')



plt.show()
```

-> One can use:
    
    1. hexadecimal color values, #4CAF50abs
    
    2.color supported names like 'hotpink', 'chartreuse', 'chocolate', 'cyan', 'darkblue'
    

### Line width

-> the keyword linewidth or lw is used to change the width of the line

```python

plt.plot(y, linestyle = 'dotted', color = 'hotpink', lw = '6')
plt.show()
```

## Matplotlib labels and title

-> the xlabel() and ylabel() functions are used to set label or x- and y-axis repectively.

-> the title() function is used to give our plot a name

```python

plt.plot(x_point, y_point, ls = '-', c = 'darkblue')


plt.xlabel("x-axis line")

plt.ylabel("y-axis line")

plt.title("OUR TITLE NAME")

plt.show()
```

## Properties of title(), xlabel(), ylabel()

### fontdict 

-> this is a parameter in xlabel(), ylabel(), title() that sets the font properties to them.

```python

font1 = {'family':'serif', 'color':'red', 'size': 20}

font2 = {'family':'serif', 'color': 'blue', 'size': 15}

plt.plot(x_point, y_point, ls = '-', c = 'darkblue')

plt.xlabel("x-axis line", fontdict = font2)

plt.ylabel("y-axis line", fontdict = font2)

plt.title("OUR TITLE NAME", fontdict = font1)

plt.show()
```

### loc

-> used inside title() function to  position the title

-> the legal values are: 'left', 'right' and 'center'

##  Grid Lines

-> the grid() function can be used to  add grid lines to the plot.

```python


font1 = {'family':'serif', 'color':'red', 'size': 20}

font2 = {'family':'serif', 'color': 'blue', 'size': 15}

plt.plot(x_point, y_point, ls = '-', c = 'darkblue')

plt.xlabel("x-axis line", fontdict = font2)

plt.ylabel("y-axis line", fontdict = font2)

plt.title("OUR TITLE NAME", fontdict = font1)

plt.grid()

plt.show()
```

### axis

->is a parameter used in grid() function to specify which line to display.

-> legal values are: 'x', 'y' and 'both'

## Grid line properties

-> you can set the line properties o the grid like this: grid(color='color', linestyle ='linestyle', linewidth = number)

```python
font1 = {'family':'serif', 'color':'red', 'size': 20}

font2 = {'family':'serif', 'color': 'blue', 'size': 15}

plt.plot(x_point, y_point, ls = '-', c = 'darkblue')

plt.xlabel("x-axis line", fontdict = font2)

plt.ylabel("y-axis line", fontdict = font2)

plt.title("OUR TITLE NAME", fontdict = font1)

plt.grid(axis = 'x', color = 'green', linewidth = '0.5', linestyle = '--')# display a vertical grid line

plt.show()
```
